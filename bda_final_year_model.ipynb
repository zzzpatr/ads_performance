```python
import pandas as pd
import datetime
import statsmodels.api as sm
import os
```

```python
all_count_df = pd.read_csv("all_count_df.csv")
```

### 1. 交易總金額

```python
X = sm.add_constant(all_count_df.drop(["TotalSalesAmount","sale_num","member_id","month"],axis=1), prepend=False)
X = X.loc[:, (X != 0).any(axis=0)]
mod = sm.OLS(all_count_df["TotalSalesAmount"], X)
res = mod.fit()
print(res.summary())
```

```python
pd.DataFrame(res.pvalues).sort_values(by=0).head(15)
```

### 2. 交易頻率

```python
X = sm.add_constant(all_count_df.drop(["TotalSalesAmount","sale_num","member_id","month"],axis=1), prepend=False)
X = X.loc[:, (X != 0).any(axis=0)]
X = (X//X).fillna(0)  # 因為這邊只看有交易的欄位，交易頻率直接建模會變成每個廣告的總數就是Y，改成每個廣告的數目變成 1 / 0
mod = sm.OLS(all_count_df["sale_num"], X)
res = mod.fit()
print(res.summary())
```

```python
pd.DataFrame(res.pvalues).sort_values(by=0).head(15)
```
